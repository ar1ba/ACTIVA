```python
#print(real.obs)
import scanpy as sc
import numpy as np
import matplotlib.pyplot as plt


expression = np.asarray(activa.X.sum(axis=0)).flatten()

# Get gene IDs
gene_ids = activa.var_names

# Create a dictionary to store gene IDs and their expression
gene_expression_dict = dict(zip(gene_ids, expression))

# Sort gene IDs by their expression values
sorted_genes = sorted(gene_expression_dict.items(), key=lambda x: x[1])

# Extract the lowest expressions and corresponding gene IDs
lowest_gene_ids = [gene[0] for gene in sorted_genes[:21]]  # Change 10 to the number of genes you want to plot
lowest_expressions = [gene[1] for gene in sorted_genes[:21]]  # Change 10 to the number of genes you want to plot

# Plot the lowest expressions in a line graph
plt.figure(figsize=(10, 6))
plt.plot(lowest_gene_ids, lowest_expressions, marker='o', linestyle='-')
plt.title('Lowest expressed 20 Genes')
plt.xlabel('Gene IDs')
plt.ylabel('Expression')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
